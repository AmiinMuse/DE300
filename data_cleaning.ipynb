```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for better notebook output
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("heart_disease(in).csv")
# df = df.DataFrame()
df.head()
# df.info()
```

```python
# Summary of data
# print(df.describe())
# print(df.info())

# # Check for missing values
# print(df.isnull().sum())
print(len(df))
print(df.info())
```

```python
# change age col to float and remove rows where age is less than 0 
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0.0)
df = df[df['age'] >= 0]

# drop duplicate rows 
df.drop_duplicates(inplace=True)

# Fill missing values with a specified value or a computed value
df.fillna(0, inplace=True)
```
